## Problem 1

As the edges attach to existing nodes uniformly at random, each node has the same probability of $\frac{1}{n}$.  
probability that nodes transfer into with degree q and transfer out from degree q:  
$$
q-1 \rightarrow q: cP_{q-1}(n)  \\
q \rightarrow q+1: cP_{q}(n)  \\
$$
Master equation:
$$
(n+1)P_q(n+1) = nP_q(n) + cP_{q-1}(n) - cP_{q}(n) 
$$

For node with degree 0,
$$
(n+1)P_0(n+1) = nP_0(n) + 1 - cP_{0}(n) 
$$

In the limit of large network, $n \to +\infty, (n+1)P_q(n+1) \approx nP_q(n)$,  

$$
\Rightarrow (1+c)P_q = cP_{q-1} \\ 
(1+c)P_0(n+1) = 1 \\
\Rightarrow P_q = \frac{1}{1+c}(\frac{c}{1+c})^q = Cr^q
$$
where C is a normalization constant and $r = \frac{c}{1+c}$

## Problem 2

### 2(a)
As the probability is proportional to degree k, $ P_i \propto k_i $, which can be represented as $ P_i = Ck_i $, where C is a normalization constant.

Sum up the probability of all nodes, the result should be two, since each edge has two end nodes:
$$
\sum_i P_i = \sum Ck_i = C<k> = 2
$$

And
$$
n<k> = 2m \Rightarrow <k> = \frac{2m}{n} \\
\Rightarrow C = \frac{2}{<k>} = \frac{1}{m} \\
\Rightarrow P_i = Ck_i = \frac{k_i}{m}
$$

### 2(b)  
with conclusion from 2(a), probability that nodes transfer into with degree k and transfer out from degree k:  
$$
k-1 \rightarrow k: \frac{k-1}{m}  \\
k \rightarrow k+1: \frac{k}{m}  \\
$$
Master equation: 
$$
nP_k(m+1) = nP_k(m) + nP_{k-1}(m)\frac{k-1}{m} - nP_k(m)\frac{k}{m}  \\
$$

### 2(c)
with conclusion from 2(b) and apply some transformation to it, we can get:  

$$
m[P_k(m+1) - P_k(m)] = (k-1)P_{k-1}(m) - kP_k(m) \tag{2-3}
$$

mean degree
$$
c = \frac{2m}{n} 
\Rightarrow m = \frac{cn}{2} 
\Rightarrow \frac{m}{dm} = \frac{c}{dc}
$$

The derivative of $P_k$ over $m$ can be approximated with $\Delta m = 1$:

$$
\frac{dP_k}{dm} = \lim_{\Delta m \approx 0} \frac{P_k(m+\Delta m) - P_k(m)}{\Delta m} 
\approx \frac{P_k(m+1) - P_k(m)}{(m+1) - m} = P_k(m+1) - P_k(m)

$$

$$

Substitute into equation (2-3),
$$

### 4(b)

$$
