As the distribution obeys perfect power-law, the probability density function of x, the number sold of book, can be written as follows:

$$
f(x) = Cx^{-\alpha} 
$$

Integrate over the range of $[1, x_{max}]$, where $x_{max}$ is the largest number of book sold, the result should be 1. 

$$
\int_{x=1}^{x_{max}} f(x)dx = \int_{x=1}^{n} Cx^{-\alpha} dx = \frac{C}{\alpha-1}[1-x_{max}^{1-\alpha}] = 1 \\
\Rightarrow C = \frac{\alpha-1}{1-x_{max}^{1-\alpha}}
$$

When $n$ is large, $x_{max}^{1-\alpha} \to 0, \Rightarrow C = \alpha-1 $.   
Thus,

$$
f(x) = (\alpha-1)x^{-\alpha} 
$$

The cumulative distribution function (CDF) can be defined as:

$$
F(x) = \int_x^{+\infty} f(u)du = \int_x^{+\infty} (\alpha-1)u^{-\alpha} du = x^{1-\alpha}
$$

In order to calculated the expected value of $x_{max}$, if we draw a sample $j$ from the distribution of x, then $j$ should lie between $x$ and $x+dx$. There are $n$ possible combinations of drawing $j$ from $n$ samples,  with probability $nf(x)$; besides, there should be $(n-1)$ samples that are lower than $x_{max}$, with probability $[1-F(x)]^{n-1}$. Finally, the total prabability should be:

$$
\pi(x) = nf(x)[1-F(x)]^{n-1}
$$

The expected value of $x_{max}$ can be calculated via integration of $x\pi(x)$ over all possible range of $[1, +\infty]$:

$$
\overline{x_{max}} = \int_1^{+\infty} x\pi(x)dx = nxf(x)[1-F(x)]^{n-1} dx \\
= n\int_1^{+\infty} x^{-\alpha}(1 - x^{1-\alpha})^{n-1} dx \\
= nB(n, \frac{\alpha-2}{\alpha-1})
$$

The Beta-function for large values can be approximaetd with a power law.

$$
B(n, \frac{\alpha-2}{\alpha-1}) \approx n^{\frac{\alpha-2}{\alpha-1}} \\
\Rightarrow \overline{x_{max}} \approx n \cdot n^{\frac{\alpha-2}{\alpha-1}} = n^{\frac{1}{\alpha-1}}
$$

## Problem 6

Since the degree distribution of the Internet obeys power law $p_k \approx k^{-\alpha}$, where $\alpha \approx 2.5$

The power-law degree distribution of the Internet:

$$
p_k = \begin{cases}
0, (k=0) \\
\frac{k^{-\alpha}}{\zeta(\alpha)}, (k \geq 1)
\end{cases}
$$
where 
$$
\zeta(\alpha) = \sum_{k=1}^{\infty}k^{-\alpha}
$$

The fundamental generating functions are:

$$
G_0(z) = \sum_{k} p_k z^k = \frac{1}{\zeta(\alpha)} \sum_{k=1}^{\infty}k^{-\alpha} z^k \\
G_1(z) = \frac{G^{'}_0(z)}{G^{'}_0(1)} = 
\frac{1}{\zeta(\alpha - 1)} \sum_{k=1}^{\infty}k^{-\alpha + 1} z^{k-1} 
$$

As $\alpha$ is assumed to be 2.5

$$
G_0(z) = \frac{1}{\zeta(2.5)} \sum_{k=1}^{\infty}k^{-2.5} z^k \\
G_1(z) = \frac{1}{\zeta(1.5)} \sum_{k=1}^{\infty}k^{-1.5} z^{k-1} 
$$

Assumptions
- The Internet has a “pure” power law degree distribution.
- The model can be approximated for the Internet, in which there is no multi-edge or self loop.

In practice
- The power law degree distribution is not ideal.
